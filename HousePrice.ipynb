```python
import pandas as pd
import numpy as np
import os
import re
```

```python
file_list = ['A_lvr_land_A.csv',
             'B_lvr_land_A.csv',
             'D_lvr_land_A.csv',
             'E_lvr_land_A.csv',
             'F_lvr_land_A.csv',
             'H_lvr_land_A.csv',
             'J_lvr_land_A.csv',
             'O_lvr_land_A.csv',]
for file in file_list:
    all_df = None
    for s in ['108-1', '108-2', '108-3', '108-4', '107-1', '107-2', '107-3', '107-4']:
        print('data/'+s+'/'+file)
        df = pd.read_csv('data/'+s+'/'+file, encoding='utf-8')
        if all_df is None:
            all_df = df
        else:
            all_df = pd.concat([all_df, df], axis=0).reset_index(drop=True)
    all_df = all_df.drop([0], axis=0).reset_index(drop=True)
    all_df = all_df.dropna(subset=['單價元平方公尺']).reset_index(drop=True)
    all_df = all_df.loc[all_df['交易標的'].str[:2] == '房地'].reset_index(drop=True)
    all_df.單價元平方公尺 = all_df.單價元平方公尺.astype('float')
    all_df.總價元 = all_df.總價元.astype('float')
    all_df.to_csv('data/108_'+file, index=False, encoding='utf-8-sig')
```

```python
file_list = ['台中房價.csv',
             '台南房價.csv',
             '新竹房價.csv',
             '桃園房價.csv',
             '雙北房價.csv',
             '高雄房價.csv']

for file in file_list:
    df = pd.read_csv('data/'+file, encoding='utf-8')
    area_list, price_list = [], []
    for area in df['鄉鎮市區'].unique():
        area_list.append(area)
        if df.loc[df['鄉鎮市區']==area].shape[0] <= 3: #濾掉交易量太少的區域(2年內<=3筆)
            price_list.append(0)
        else:
            price_list.append(df.loc[df['鄉鎮市區']==area].單價元平方公尺.mean())
        area_df = pd.DataFrame({'area2':area_list, 'mean_price':price_list})
        area_df.to_csv('processed/'+file, index=False, encoding='utf-8-sig')
        print('Area: %s | Average price:%f'%(area, df.loc[df['鄉鎮市區']==area].單價元平方公尺.mean()))
```
